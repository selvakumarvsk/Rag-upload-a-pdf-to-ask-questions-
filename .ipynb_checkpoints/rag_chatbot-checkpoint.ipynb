```python
import langchain
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

load_dotenv()

google_api_key = os.getenv("GOOGLE_API_KEY")

google_llm = ChatGoogleGenerativeAI(
    model = "gemini-3.1-flash-lite",
    temperature = 0,
    api_key = google_api_key,
)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system","You are a helpfull assistent ansewer the questions"),
    ("human","summerise about this person {input}")
])

# chain = prompt | google_llm | StrOutputParser()

# res = chain.invoke({"input":"trump"})
# print(res)
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("resume.pdf")
documents = loader.load()

print(documents[0].page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitters = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 50
)

chunks = text_splitters.split_documents(documents)

print(len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2"
)
```

```python
print(embeddings.embed_query("hello"))
```

```python
from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_documents(chunks,embeddings)
```

```python

vectorstore.save_local("faiss_index")
```

```python
res = vectorstore.similarity_search("who is selva?")
print(res[0].page_content)
```

```python
retrival = vectorstore.as_retriever()
print(retrival.invoke("Who is selva"))
```

```python
from langchain_core.runnables import RunnablePassthrough

prompt = ChatPromptTemplate.from_template(
    """Answer the question using the context:
    context: {context}
    question: {question}
    
    answer: """
)

chain = {"context":retrival, "question":RunnablePassthrough()} | prompt | google_llm | StrOutputParser()

print(chain.invoke("tell me about selva is that he is perfect for this role"))
```
